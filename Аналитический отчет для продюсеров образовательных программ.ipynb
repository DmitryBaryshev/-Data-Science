```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_courses = pd.read_csv('courses.csv', index_col= 'Unnamed: 0')
df_progresses = pd.read_csv('progresses.csv')
df_course_cont = pd.read_csv('course_contents.csv')
df_progresses_phases = pd.read_csv('progress_phases.csv')
df_students = pd.read_csv('students.csv')
```

```python
main = df_progresses.merge(df_courses, left_on='course_id', right_on='id')
main.rename(columns={'id_x': 'progress_id', 'title': 'course_title', 'field': 'course_field'}, inplace=True)
main.drop(['id_y'], axis=1, inplace=True)
```

```python
main.head()
```

```python
main = main.merge(df_course_cont)
main = main.merge(df_progresses_phases,left_on=['progress_id', 'module_number', 'lesson_number'], right_on=['progress_id', 'module_number', 'lesson_number'])
```

```python
main[main.student_id == '768c2987a744c51ce64a5993a2a94eaf'].head()
```

Количетсво курсов в датасете я определил следующим образом:

```python
main['course_id'].drop_duplicates().count()
```

Количество модулей на каждом курсе,

```python
main.groupby('course_title')['module_title'].apply(lambda x: x.unique().size).sort_values()
```

Количество уроков в каждом модуле на каждом курсе

```python
main.groupby(['course_title', 'module_title'])['lesson_title'].aggregate(lambda x: x.unique().size).reset_index().sort_values(['course_title', 'lesson_title'], ascending=(True, False)).head()
```

Медианное количество уроков в модуле на каждом курсе

```python
main.groupby(['course_title', 'module_title'])['lesson_title'].aggregate(lambda x: x.unique().size).groupby(['course_title']).aggregate(np.median)
```

Количество учеников на каждом курсе

```python
main.groupby(['course_title'])['student_id'].apply(lambda x: x.unique().size)
```

Минимальный, максимальный, средний, медианный возраст студентов

```python
fig, ax = plt.subplots()
df_chart= main.pivot_table(values='student_id', index='course_title', columns='course_field', aggfunc=lambda x: x.unique().size, margins=True).sort_values('All', ascending=True).drop(index=['All'], columns=['All'])


df_chart.plot.barh(ax=ax, title='Количество студентов на курсах', stacked=True, alpha=0.1)


df_chart = df_chart.stack().reset_index()
ax.vlines(df_chart[0].median(), ax.get_ylim()[0], ax.get_ylim()[1], color='r', linestyles='dashed', linewidth=1)
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.set(xlabel='', ylabel='')

plt.show()
```

1. (p) Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

```python
# Формирование датасета через группировку по модулю и студенту
# с определением дат начала и окончания всех домашних заданий в модуле
df_conversion = main.loc[
    main['is_homework']==True, 
    ['course_title', 'course_field', 'module_number', 'module_title', 'student_id', 'start_date', 'finish_date']
].groupby(
    ['course_title', 'course_field', 'module_number', 'module_title', 'student_id']
).aggregate(
    {'start_date': np.min, 'finish_date': np.max}
)

# Перегруппировка датасета по модулю c подсчетом
# количества студентов начавших и завершивших обучение 
df_conversion = df_conversion.assign(
    student_start=df_conversion['start_date'].notnull(), student_finish=df_conversion['finish_date'].notnull()
).groupby(
    ['course_title', 'course_field', 'module_number', 'module_title']
).aggregate(
    {'student_start': np.sum, 'student_finish': np.sum, 'start_date': np.min, 'finish_date': np.max}
).sort_index().reset_index()

# Преобразование формата дат к краткому виду
df_conversion['start_date'] = df_conversion['start_date'].astype('datetime64[D]')
df_conversion['finish_date'] = df_conversion['finish_date'].astype('datetime64[D]')

# Расчет конверсии и добавление значений в датасет
for course in df_conversion['course_title'].unique():
    for index, module_number in df_conversion.loc[df_conversion['course_title']==course, 'module_number'].items():
        # Определение количества студентов, начавших обучение в текущем модуле
        student_start = df_conversion.loc[index, 'student_start']
        # Определение количества студентов, завершивших обучение в предыдущем модуле
        student_finish = df_conversion.loc[
            (df_conversion['course_title']==course) & (df_conversion['module_number'] < module_number), 'student_finish'
        ].tail(1).min()
        # Подсчет конверсии текущего модуля
        df_conversion.loc[index, 'conversion'] = student_start / student_finish if student_finish > 0 else 0

# Вывод первых 5-ти позиций
df_conversion.head()
```

```python
# Определение основных объектов диаграммы
fig, ax = plt.subplots(figsize=(26, 10))

# Вывод диаграммы на страницу
df_conversion.pivot_table(
    values='conversion', index=df_conversion.index, columns='course_title'
).plot.bar(
    ax=ax, title='Конверсия перехода студентов из одного модуля в другой на курсах', stacked=True
)

# Вывод подписей делений оси абсцисс
ax.set_xticklabels(df_conversion['module_title'].apply(lambda x: f'{x[:30]}...'), rotation=45, ha='right')

# Вывод подписей осей координат
ax.set_xlabel('Модули курса')
ax.set_ylabel('Конверсия')

# Автоматическое позиционирование элементов диаграммы
fig.tight_layout()

# Отображение диаграммы на странице
plt.show()
```
